# CRIMES AGAINST WOMEN IN INDIA (2001-2021)
 Crimes against women in India represent a grave societal crisis, highlighting entrenched gender disparities and systemic challenges. Despite advancements in laws and heightened public awareness, the reality of women's safety remains distressing. These crimes span across various forms, including domestic abuse, sexual violence, dowry deaths, human trafficking, and honor killings, each underscoring the pervasive violence women face in both public and private spheres. The impact is not confined to individual suffering but reverberates through families, communities, and the fabric of society as a whole.

### The Importance of Data-Driven Solutions
A data-driven approach is essential to comprehensively address and combat these crimes. By meticulously analyzing crime data, stakeholders can uncover critical patterns, identify high-risk areas, and gauge the effectiveness of legal and social interventions. This approach goes beyond surface-level observations, helping to pinpoint systemic failures and cultural factors that perpetuate violence. It empowers decision-makers to formulate targeted policies, track progress, and ensure accountability. Data analysis also enables early warning systems, facilitating timely interventions in regions or demographics particularly vulnerable to gender-based violence.

```python
crime = data.groupby('Year').sum()

plt.figure(figsize=(15, 8))
sns.lineplot(data=crime)
plt.title('Crimes Against Women (2001-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')

# Ensure there is a continuous range of years
years = list(range(2001, 2022)) 
plt.xticks(ticks=years, rotation=45)

plt.legend(title='Crime Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()
```

```python
data.groupby(['Year']).agg({
    'Rape': 'sum',
    'Kidnapping and Abduction': 'sum',
    'Dowry Deaths': 'sum',
    'Assault against Women': 'sum',
    'Assault against Modesty of Women': 'sum',
    'Domestic Violence': 'sum',
    'Women Trafficking': 'sum'
}).plot(kind='bar', figsize=(23,10))
plt.title('Crime Proportion by Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.legend(title="Crime Type")  # Adding a legend to identify the crime types
plt.show()
```
